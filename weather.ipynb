```python
import requests
import datetime
import streamlit as st
```

```python
st.title('WEATHER APP')
city=st.text_input("Enter city name")
api="ed9467182326e5d1be9d18dc60937497"
IST_OFFSET=datetime.timezone(datetime.timedelta(hours=5,minutes=30))
if st.button("Get Weather"):
  url=f"https://api.openweathermap.org/data/2.5/weather?q={city}&appid={api}&units=metric"
  response=requests.get(url)
  if response.status_code==200:
      data=response.json()
      st.success(f"Weather in {city}")
      st.metric("Temperature:", f"{data['main']['temp']} °C")
      st.metric("Humidity:", f"{data['main']['humidity']}%")
      st.metric("Wind Speed",f"{data['wind']['speed']} m/s")
      sunrise_t=data['sys']['sunrise']
      sunset_t=data['sys']['sunset']
      sunrise=datetime.datetime.fromtimestamp(sunrise_t,tz=IST_OFFSET).strftime('%I:%M %P')
      sunset=datetime.datetime.fromtimestamp(sunset_t,tz=IST_OFFSET).strftime('%I:%M %P')
      st.write('Sunrise:',sunrise)
      st.write('Sunset:',sunset)
  else:
     st.write("city is not found ")
```
